# GNU Plot Cheat sheet

This file contains various samples that would be reused this term.

```gnuplot
reset
set lmargin 11 # margin on the left
 # draw lines for each ytics and mytics

set terminal pngcairo size 1000,400 # pngcairo!

set grid ytics
set key right top
set key outside

set xlabel "time"
set ylabel "velocity" rotate by 0

g(x) = cos(a*x)/a
plot a=1 g(x) title "plot1",\
    a=2 g(x) title "plot2"
```

```gnuplot
reset
set terminal pngcairo size 400,400
set style line 1 \
    linecolor rgb '#0060ad' \
    pointtype 7 pointsize 1.5

set xrange [0:5]
set yrange [-1:4]

plot '/tmp/sample.dat' with linespoints linestyle 1 title "Data"
```

```gnuplot
reset
set terminal pngcairo size 1000,400
set grid
splot x**2+y**2, x**2-y**2
```
